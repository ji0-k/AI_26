```python
# 머신러닝 알고리즘은 지도학습, 비지도학습(벽돌깨기)으로 나누어짐
# 지도학습 : 훈련하기 위한 데이터와 정답이 필요함(전처리데이터가 중요)
# 지도학습에서 데이터와 정답을 입력(input)과 타겟(target)이라고 함
# 이 둘을 합쳐서 훈련데이터(training data)라고 함
# 입력으로 사용된 길이와 무게 -> 특성(feature)

# MBC마켓에서 살아있는 생선을 최초로 판매하려고 한다.
# 고객이 온라인 주문을 하면 가장 빠른 물류 센터에서 신선한 생선을 곧바로 배송
# 여기서 문제 발생 : 생선 포장 직원이 생선의 종류를 알지 못함!!!
# IP 카메라를 이용해서 수족관을 실시간 촬영을 하고 이것을 객체추적 기술을 사용

# 우선 도미와 빙어 2개의 객체를 찾도록 훈련시작!!!!
# 1. 전처리 데이터가 필요함 : 도미 생선의 길이, 무게를 학습을 시켜보자.
```

```python
# 도미 생선의 길이 : bream_length 도미용 길이 리스트
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]

# 도미 생선의 무게 : bream_weigh 도미용 무게 리스트
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
# 특성(feature) : 데이터의 특징
# 첫번째 도미 특성 ( 길이 : 26.4cm, 무게 : 242.0g)
# 두번째 도미 특성 ( 길이 : 26.3cm, 무게 : 290.0g)

# 두 특성을 숫자로 보는 것 보다 그래프로 표현하면 데이터를 잘 이해 할 수 있다.
# 길이 : X축 , 무게 : y축 -> 산점도 ( 두 좌표와의 관계 )
```

```python
import matplotlib.pyplot as plt
# 파이썬에서 과학 계산용 그래프를 그리는 대표적인 패키지 : matplotlib
#https://wikidocs.net/92071 사용법

plt.scatter(bream_length, bream_weight) # 차트에 값을 넣음
#   scatter 차트의 종류 (X축 : bream_length, Y축 : bream_weight)
plt.xlabel('length')          # X축의 이름표
plt.ylabel('weight')          # Y축의 이름표
plt.show()
# 도미 35마리를 2차원 그래프에 점으로 나타내봄
# 생선의 길이가 길수록 무게가 많이 나간다~!!! -> 선형적이라고함
```

```python
# 빙어 데이터 14개 준비

smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
plt.scatter(smelt_length, smelt_weight) # 차트에 값을 넣음
#   scatter 차트의 종류 (X축 : bream_length, Y축 : bream_weight)
plt.xlabel('smelt_length')          # X축의 이름표
plt.ylabel('smelt_weight')          # Y축의 이름표
plt.show()
```

```python
# 이 2개의 리스트를 합쳐서 ai 학습용 자료로 만들어야 함

length = bream_length + smelt_length     # 도미길이리스트 + 빙어길이리스트 합치기
weight = bream_weight + smelt_weight     # 도미무게리스트 + 빙어무게리스트 합치기

# 사이킷런을 이용한 ai트레이닝 시작!!!
# 사이킷런의 입력은 2차원 리스트로 만들어야 함

print(length)                            # 합쳐진 전체 길이 리스트 출력
print(weight)
```

```python
# zip() 나열된 리스트 각각에서 하나씩 원소를 꺼내어 반환
fish_date = [[l,w] for l,w in zip(length, weight)]
# length, weight를 2차원 리스트로 만듬
print(fish_date) # 2차원리스트 출력 해봄
```

```python
# 머신러닝은 정답을 target 이라는 값으로 알려줘야 한다
# target 0 (빙어), 1(도미) 처리하려함

# 생선 49개에 대한 정답 리스트 만들어줘야함
# 머신러닝 에서는 정답 데이터를 2진분류로 테스트 해볼까 함
fish_target = [1]*35 + [0]*14
print(fish_target)
```

```python
# pip install scikit-learn 파이참에서는 설치 후 구동
# 전처리 데이터가 완성 되었으니 사이킷런 패키지를 활용해서 트레이닝을 해보자

from sklearn.neighbors import KNeighborsClassifier # 사이킷 패키지 안에. 이웃패키지
#사이킷런패키지.이웃모듈      KNeighborsClassifier k-최근접이웃알고리즘
#                             2진분류용 클래스

kn = KNeighborsClassifier() # import된 객체를 생성하여 변수에 연결
# kn.하면 k-최근접 알고리즘을 사용한다.
# k는 기본값 5개이다.
# kn.fit() 훈련시작
# kn.socre() 평가 -> 점수보기
# kn.predict() 결과 -> 샘플을 넣어서 2진 비교
```

```python
# fish_data, fish_target을 전달하여 도미를 찾기 위한 기준을 학습(.fit) 시킨다
# (트레이닝데이터 : 훈련정답데이터)

kn.fit(fish_date, fish_target) # 훈련 끝
```

```python
# kn 얼마나 잘 훈련되었는지 평가를 해보자 score()
kn.score(fish_date, fish_target) # 1.0 -> 정확도100%
# 현재 더미데이터가 많지 않아 훈령용과 평가용 데이터가 같음으로 100%결과가 나옴
# 차후에는 더미 데이터를 쪼개서 평가하는것이 좋다
```

```python
# k- 최근접 이웃 알고리즘 파악 해보기
# 어떤 데이터에 대한 답을 구할 때 주위의 다른 데이터를 보고
# 다수를 차지하는 것을 정답으로 출력
```

```python
kn.predict([[30,600]]) #.predict[길이.무게] 새로운 데이터의 정답을 예측 (2진분류)
# 결론 1-> 도미
```

```python
kn.predict([[8,10]]) #.predict[길이.무게] 새로운 데이터의 정답을 예측 (2진분류)
# 결론 0-> 빙어
```

```python
# k - 최근접 이웃 알고리즘을 위해 준비해야 할 일은 데이터를 모두 가지고있는 게 전부
# 새로운 데이터에 대해 예측할 때는 가장 가까운 직선거리에 어떤 데이터가 있는지 살치기만 함
# 단점. 데이터가 아주 많은 경우 메모리가 많이 필요, 계산 시간도 오래걸리므로 사용하기 어렵다
# 객체를 메모리에 만들 고 활용한다. -> w재부팅 시 fit 다시해야함
print(kn._fit_X) # _fit__X 내장된 변수로 훈련 데이터를 보관함
```

```python
print(kn._y) # 정답파일 1: 도미 0: 빙어
```

```python
# 결론 : k-최근접 이웃 알고리즘은 훈련결과 파일이 존재하지않는다 (메모리에 로드->재부팅시 사라짐)
# 새로운 데이터가 들어오면 가장 가까운 데이터를 메모리에서 참고하여 도미, 빙어를 구분함
kn.predict([[5,15]]) # 길이가 5cm, 무게가 15g 인값을 판단 -> 0:빙어
```

```python
#객체 생성시 n_neighbors = ? 참고 할 이웃 데이터 수 기본 값 5
# 이웃 객체 5개 찾아서 결론 냄
# 만약 5개가 아닌 다른값으로변경 하여 데이터 훈련 하고싶으면

kn49 = KNeighborsClassifier(n_neighbors= 49) # 참고 데이터 49개로 한 모델 kn49변수
#      KNeighborsClassifier(n_neighbors= 5) 기본

#원본타겟, 원본데이터가 49인데 가장 가까운 49개 (전체 데이터가 이웃)

kn49.fit(fish_date,fish_target)
kn49.score(fish_date,fish_target) # 71%, 좋지않은 성능을 보임
```

```python
# n_neighbors= 값을 변경 해 가면서 최적의 정답을 찾는것이 머신러닝 훈련 개발자의 목표
kn49.predict([[5,15]]) #r 길이 5cm, 무게 15g인 데이터 찾기
```
