```python
import numpy as np

ndarray_1 = np.array([1,2,3])
ndarray_2 = np.array([4,5,6])
list_1 =[1,2,3]
list_2 =[4,5,6]

print(ndarray_1 +ndarray_2) # 넘파이 배열 힙
print(list_1 + list_2)      # 리스트 배열 합 결과 다름
```

```python
result_list = [x+y for x,y in zip (list_1, list_2)]
result_list2 = [x+y for x,y in zip (list_1, ndarray_1)]
print(result_list)
print(result_list2)
```
