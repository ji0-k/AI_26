```python
# 전에 농어 무게를 예측하는 k-최근접알고리즘을 사용해서
# 회귀 모델을 사용해본 결과 정확한 예측이 어려울 수 있다라는 결론이 생김
# (원본데이터의 개수 부족)

# 길이가 50cm인 농어가 들어왔을 때의 회귀는??? (한계점이 생김), 기존 학습데이터 없는 범위
```

```python
import numpy as np
# 넘파이를 이용해서 배열을 생성함
# 농어 56개 데이터임
perch_length = np.array([8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 21.0,
       21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 22.5, 22.7,
       23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 27.3, 27.5, 27.5,
       27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 36.5, 36.0, 37.0, 37.0,
       39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 40.0, 42.0, 43.0, 43.0, 43.5,
       44.0])
perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])
```

```python
# 왜 reshape(-1, 1)을 해야 할까요?
# 사이킷런(sklearn)의 모델들은 데이터를 입력받을 때 **"행(Row)은 샘플, 열(Column)은 특성"**인 2차원 배열 형태를 기대
# 현재 데이터: [1, 2, 3] (1차원 리스트, 그냥 숫자 나열)
# 필요한 데이터: [[1], [2], [3]] (2차원 행렬, "1번 물고기의 길이는 1이다"라는 구조)
# 1 (두 번째 인자)"열(특성)의 개수를 1개로 고정해라"
# -1 (첫 번째 인자)"행의 개수는 전체 원소 개수에 맞춰서 네가 알아서(자동으로) 계산해라"

from sklearn.model_selection import train_test_split

# 데이터를 훈련 세트와 테스트 세트로 섞어서 나눕니다.
train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight, random_state=42)
# perch_length(입력), perch_weight(정답)를 넣으면 4개로 쪼개져서 나옵니다.

# [중요] 사이킷런 모델은 입력 데이터가 '2차원 배열'이어야 학습이 가능합니다.
# 현재 train_input은 1차원 배열(예: [10, 12, 11...]) 상태입니다.

# train_input을 (샘플 개수, 1) 형태의 2차원으로 바꿉니다.
# -1은 '남은 자리를 알아서 채워라', 1은 '열의 크기를 1로 해라'는 명령입니다
train_input = train_input.reshape(-1,1)

# 테스트 세트도 모델에 넣어서 검증해야 하므로 똑같이 2차원으로 바꿔줍니다.
test_input = test_input.reshape(-1,1)

# 변환 후 shape 확인 (예: (36,) -> (36, 1))
print(train_input.shape, test_input.shape)
```

```python
# 회귀 알고리즘용 구현 클래스
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor(n_neighbors = 3)
# 모델 객체 생성 , 가장 가까운 데이터 3개 참고
# 사이킷 런의 k-최근접이웃 기본값이 5인데 이것을 3으로 변경

knr.fit(train_input, train_target)
#모델 훈련 , 데이터 , 무게정답

print(knr.predict([[50]]))
# 새로운 데이터(길이 50cm)를 넣어 무게를 예측합니다.
# 입력값이 [[50]]인 이유는 모델이 2차원 배열 형태를 기대하기 때문입니다
```

```python
# k-최근접이웃 회귀는 "평균"을 내는 방식
# 가장 가까운 이웃 3개를찾아 평균을 찾아...데이터 범위 밖의 값 예측 오류 -> k최근접이웃회귀한계
# 해결 방안 : 선형회귀 Linear Regression , 하나의 직선(y = ax + b)

# n_neighbors의 영향
# 값을 너무 작게 잡으면(예: 1): 데이터 하나하나의 변동에 너무 민감하게 반응합니다 (과대적합 위험).
# 값을 너무 크게 잡으면(예: 전체 데이터 개수): 모든 데이터의 평균을 내버리므로 모델이 너무 단순해집니다 (과소적합 위험).
# 그래서 보통 기본값인 5를 쓰거나, 지금처럼 적절히 조절하며 테스트합니다.

```

```python
#k-최근접 이웃 한계 시각적 증명

import matplotlib.pyplot as plt

distances, indexes = knr.kneighbors([[50]])
# 50cm 농어 이웃들을 찾아 거리(distances)와 인덱스(indexes)를 가져옵니다.

plt.scatter(train_input, train_target, label='Training Data')
# 전체 훈련 데이터를 산점도로 그립니다.


plt.scatter(train_input[indexes], train_target[indexes], marker='D')
# 훈련 세트 중에서 이웃 3가지 샘플만 다시 그림        #다이아몬드표

plt.scatter(50, 1033, marker='^')
                      #삼각형표시
plt.xlabel('length')
plt.ylabel('weight')
plt.legend() # 범례를 추가
plt.show() # 결론 50이상의 길이 데이터가 없고 무게도 없다 (아래쪽의 평균이 나옴)
```

```python
print(np.mean(train_target[indexes])) # 45센티 근방으로 평균값이 나옴
print(knr.predict([[100]]))     # 100센티짜리로 결과
```

```python
# 결론 기본데이터 부족으로 한계가 생김

# 100cm 농어의 이웃을 구합니다
distances, indexes = knr.kneighbors([[100]])

# 훈련 세트의 산점도를 그립니다
plt.scatter(train_input, train_target)
# 훈련 세트 중에서 이웃 샘플만 다시 그립니다
plt.scatter(train_input[indexes], train_target[indexes], marker='D')
# 100cm 농어 데이터
plt.scatter(100, 1033, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()  # k최근접 이웃 알고리즘은 가볍고 사용하기 쉬운 대신 한계가 있다.
```

```python
# k-최근점 이웃 알고리즘의 한계에 대안 -> 선형회귀
```

image.png

```python
# 선형회귀(linear:리니어) 비교적 간단하고 성능이 뛰어남
# 선형이라는 말에서 짐작할 수 있듯이 직선을 학습하는 알고리즘
# 빨간선이 가로로 되어 있으면 농어의 무게를 1개로 인식함
# 빨간선이 위에서 시작해서 아래로 떨어지면 반대로 예측
# 빨간선이 아래에서 시작해서 위로 올라가면 정상적인 차트로 구현

# 싸이킷런에서 제공하는 클래스를 사용해 보자
from sklearn.linear_model import LinearRegression
# LinearRegression 선형 회귀용

lr = LinearRegression() # 선형 회귀용 객체 생성하여 변수에 연결
lr.fit(train_input, train_target)   # 선형 회귀용으로 훈련 시작
```

```python
print(lr.predict([[50]]))
# 농어의 무게 기울기+절편 (y(농어의 무게) = a * x(농어의 길이) + b)
# 직선의 방정식!!! 무게(1241.83g)로 예측
```

image.png

```python
print(lr.coef_, lr.intercept_) # lr.coef_ 계수, lr.intercept_ 인터셉트
# 기울기를 종종 계수(coefficent), 가중치(weight)라고도 부른다.
```

```python
# 그래프로 확인 해보자.

# 훈련 세트의 산점도를 그립니다
plt.scatter(train_input, train_target)
# 15에서 50까지 1차 방정식 그래프를 그립니다
plt.plot([15, 50], [15*lr.coef_+lr.intercept_, 50*lr.coef_+lr.intercept_])
# 직선차트 x축값                 y축값[15*계수+인터셉트, 50*계수+인터셉트]
#                농어의 무게공식 = 기울기*농어의길이+절편

# 50cm 농어 데이터
plt.scatter(50, 1241.8, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 결론 예측은 가능한데 직선이라 값이 너무 직선적이다.

print(lr.score(train_input, train_target))  # 93.98%
print(lr.score(test_input, test_target))    # 82.47%
# 훈련세트가 테스트에 비해 높지만 차이가 많이 남으로 과소적합이라고 함 (5%내외가 좋다.)
```

```python
# 다항회귀(Polynomial Regression) 제곱을 활용해서 직선을 곡선으로 처리가능
```

image.png

```python
# 다항회귀는 2차원 방정식의 그래프로 그리려면
# 길이를 제곱한 항이 훈련세트에 추가되어야 함

# 선형회귀의 문제점은 왼쪽 하단에 있는 기준점이 불일치가 된다.
# -> 곡선으로 만들어야 함
train_poly = np.column_stack((train_input ** 2, train_input))
# 제곱한 것과 두 배열을 나란히 붙인다.
test_poly = np.column_stack((test_input ** 2, test_input))
print(train_poly.shape, test_poly.shape)
# 2개의 특성이 만들어짐 (제곱값,원본값)
```

```python
print(train_poly)
```

```python
lr = LinearRegression()             # 선형회귀 객체 생성
lr.fit(train_poly, train_target)    # 훈련시작!

print(lr.predict([[50**2, 50]]))    # 결과 테스트
# 테스트할 때 이 모델의 농어 길이의 제곱과 원래 길이를 함께 넣음
# 결과 50센티의 무게는 1573.98g로 나옴
# 직선형에서는 50센티의 무게는 1241.83g으로 나옴
```

```python
# 농어의 무게 기울기+절편
# (y(농어의 무게) = a * x(농어의 길이) + b)
print(lr.coef_, lr.intercept_) # lr.coef_ 계수, lr.intercept_ 인터셉트
# 1573.98 = 1.01         * 50**2            - 21.55 * 50 + 116.05
# 무게 = lr.coef_[0] * 농어의길이 **2 - lr.coef_[1] * 50 + lr.intercept_
```

image.png

```python
# 구간별 직선을 그리기 위해 15에서 49까지 정수 배열을 만듭니다
point = np.arange(15, 50)
# 훈련 세트의 산점도를 그립니다
plt.scatter(train_input, train_target)
# 15에서 49까지 2차 방정식 그래프를 그립니다
plt.plot(point, 1.01*point**2 - 21.6*point + 116.05) # 다항회귀 공식 적용
# 50cm 농어 데이터
plt.scatter([50], [1574], marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target))

# 문제 해결을 위해 선형회귀와 다항회귀를 통해 문제를 해결하려고 했다.
# 하지만 아직 모델에 과소적합이 남아있다.
# 이를 해결하기 위해 조금 더 복잡한 모델을 만들어야 할 것 같다.
# 또 복잡한 모델을 만들다가 너무 복잡해진 경우
# 모델의 복잡도를 억제하는 방법도 알아봐야 할 것 같다.
```

```python
# 과대적합 (Overfitting)
# 모델이 훈련 세트에만 너무 집착해서, 훈련 데이터는 완벽하게 맞히지만 테스트 세트에서는 엉터리 결과를 내는 상태입니다.
# 증상: 훈련 세트 점수는 매우 높은데, 테스트 세트 점수는 현저히 낮음.
# 원인: 모델이 너무 복잡함(예: K-NN에서 이웃 개수 n_neighbors를 너무 작게 잡음).

# 과소적합 (Underfitting)
# 모델이 너무 단순 훈련 세트 제대로 학습하지 못하거나, 훈련 세트보다 테스트 세트 점수가 더 높은 상태입니다.
# 증상: 훈련 세트와 테스트 세트 점수가 둘 다 낮거나, 테스트 세트 점수가 더 높음.
# 원인: 모델이 너무 단순함(예: 데이터는 곡선인데 직선으로만 예측함), 또는 데이터 양이 너무 적음.
```
